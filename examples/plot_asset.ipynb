```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
files=['a2c_no_tur.csv','ppo_no_tur.csv','dji.csv','a2c_tur.csv','ppo_tur.csv']
colors = ['r','y','b','g','c','m', 'k']


a2c_cumpod_no_tur = pd.read_csv(files[0])
ppo_cumpod_no_tur = pd.read_csv(files[1])
dji_cumpod = pd.read_csv(files[2])
a2c_cumpod_tur = pd.read_csv(files[3])
ppo_cumpod_tur = pd.read_csv(files[4])
# print(a2c_cumpod)
plt.plot(a2c_cumpod_no_tur['daily_return'],'{}-'.format(colors[0]), label=files[0],linewidth=1)
plt.plot(ppo_cumpod_no_tur['daily_return'],'{}-'.format(colors[1]), label=files[1],linewidth=1)
plt.plot(dji_cumpod['daily_return'],'{}-'.format(colors[2]), label=files[2],linewidth=1)
plt.plot(a2c_cumpod_tur['daily_return'],'{}-'.format(colors[3]), label=files[3],linewidth=1)
plt.plot(ppo_cumpod_tur['daily_return'],'{}-'.format(colors[4]), label=files[4],linewidth=1)
plt.legend()
plt.savefig('result.png')
```
